```python
def recopilador():
    datos = []
    dato =()
    k=0
    for a in range(392):
        url = 'https://api.harvardartmuseums.org/person?q=gender:*&size=100&page={}&apikey=90d53660-6a9e-11e9-9b9e-216b7438ac22'.format(k)
        k+=1
        response = requests.get(url)
        dato = response.json()
        datos.append(dato)
    return datos
```

```python
datos=recopilador()
```

```python
df2= pd.DataFrame()
for a in datos:
    data = a['records']
    df = pd.DataFrame(data)
    df2=pd.concat([df,df2],axis=0,ignore_index=True)
```
